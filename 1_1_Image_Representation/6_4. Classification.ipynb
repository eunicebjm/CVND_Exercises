# Feature Extraction

Create a feature that represents the brightness in an image. We'll be extracting the **average brightness** using HSV colorspace. Specifically, we'll use the V channel (a measure of brightness), add up the pixel values in the V channel, then divide that sum by the area of the image to get the average Value of the image.


---
###  Find the average brightness using the V channel

This function takes in a **standardized** RGB image and returns a feature (a single value) that represent the average level of brightness in the image. We'll use this value to classify the image as day or night.

````python
# Find the average Value or brightness of an image
def avg_brightness(rgb_image):
    # Convert image to HSV
    hsv = cv2.cvtColor(rgb_image, cv2.COLOR_RGB2HSV)

    # Add up all the pixel values in the V channel
    sum_brightness = np.sum(hsv[:,:,2])
    area = 600*1100.0  # pixels
    
    # find the avg
    avg = sum_brightness/area
    
    return avg
````

# Classification and Visualizing Error

In this section, we'll turn our average brightness feature into a classifier that takes in a standardized image and returns a `predicted_label` for that image. This `estimate_label` function should return a value: 0 or 1 (night or day, respectively).

````python
# DETERMINING THE BRIGHTNESS THRESHOLD
# check the brightness levels in day photos vs night photos

day_brightness = list()
night_brightness =  list()
for img in STANDARDIZED_LIST:
    img_array = img[0]
    label = img[1]
    brightness = avg_brightness(img_array)
    
    if label == 0: #night
        night_brightness.append(brightness)
    else:
        day_brightness.append(brightness)

print(len(day_brightness))
print(len(night_brightness))
````

````python
# This function should take in RGB image input
def estimate_label(rgb_image, threshold):
    
    # extract average brightness feature from an RGB image 
    img_array = rgb_image[0]
    # label = img[1]
    brightness = avg_brightness(img_array)
    
    # set the value of a threshold that will separate day and night images
    # threshold = 101
    
    # return the predicted_label (0 or 1) based on whether the avg is 
    # above or below the threshold
    predicted_label = 0
    if brightness > threshold:
        predicted_label = 1

    return predicted_label    
    
````

````python
## Test out your code by calling the above function and seeing 
# how some of your training data is classified

def count_incorrect(threshold):
    incorrect_classifications = list()
    for i, img in enumerate(STANDARDIZED_LIST): 
        true_label = img[1]
        predicted_label = estimate_label(img,threshold)
        if true_label != predicted_label:
            incorrect_classifications.append((img, predicted_label, true_label))
    print("{} out of {} images incorrectly classified".format(len(incorrect_classifications),len(STANDARDIZED_LIST)))
    return incorrect_classifications
````

````python
count = len(STANDARDIZED_LIST)
best_threshold = 0
for threshold in range(94, 120):
    incorrect = count_incorrect(threshold)
    incorrect_count = len(incorrect)
    if incorrect_count < count:
        count = incorrect_count
        best_threshold = threshold

print("Threshold of {} produces the least wrong classifications at only {} mistakes (Accuracy of {}%)".format(
    best_threshold, count, ((len(STANDARDIZED_LIST) - count) / len(STANDARDIZED_LIST))*100
))
````
